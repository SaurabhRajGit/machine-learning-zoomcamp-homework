```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=42)
```

```python
len(df_train), len(df_val), len(df_test)
```

```python
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
y_train = df_train.y.values
y_val = df_val.y.values
y_test = df_test.y.values

del df_train['y']
del df_val['y']
del df_test['y']
```

```python
categorical = list(df.dtypes[df.dtypes == 'object'].index)
categorical
```

```python
from sklearn.metrics import mutual_info_score
```

```python
mutual_info_score(df_full_train.job, df_full_train.y)
```

```python
def mutual_info_y_score(series):
    return mutual_info_score(series, y_train)
```

```python
mi = df_train[categorical].apply(mutual_info_y_score)
mi.sort_values(ascending=False)
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression, ridge_regression
```

```python
model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
```

```python
dv = DictVectorizer(sparse=False)

train_dict = df_train[categorical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dict)
X_train
```

```python
val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
model.fit(X_train, y_train)
```

```python
model.intercept_[0]
```

```python
model.coef_[0].round(3)
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
```

```python
y_decision = (y_pred >= 0.5)
```

```python
orig_accuracy = (y_val == y_decision).mean().round(2)
orig_accuracy
```

```python
df_train[categorical + numerical].columns
```

```python
cols = categorical + numerical
cols
```

```python
for c in cols:
    # print(cols)
    # print()
    cols_train = cols.copy()
    # print()
    # print(cols_train)
    cols_train.remove(c)
    # print()
    # print(cols_train)
    dv = DictVectorizer(sparse=False)
    train_dict = df_train[cols_train].to_dict(orient='records')
    X_train = dv.fit_transform(train_dict)
    
    val_dict = df_val[cols_train].to_dict(orient='records')
    X_val = dv.transform(val_dict)
    
    model.fit(X_train, y_train)
    
    model.intercept_[0]
    
    model.coef_[0].round(3)
    
    y_pred = model.predict_proba(X_val)[:, 1]
    
    y_decision = (y_pred >= 0.5)
    
    accuracy = (y_val == y_decision).mean()
    
    print(f'without column - {c} : accuracy = {accuracy}, difference = {orig_accuracy - accuracy}')
```
